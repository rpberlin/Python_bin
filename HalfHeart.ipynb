```python
import matplotlib.pyplot as plt
import numpy as np

rad = 2
nPts = 20
ymin = -5
phi = np.arctan(rad/abs(ymin))

heartHalf1x = []
heartHalf1y = []
heartHalf2x = []
heartHalf2y = []



for i in range(0,nPts):
    theta = -2*phi + i*(np.pi+2*phi)/(nPts-1)
    heartHalf1x.append(rad + rad*np.cos(theta))
    heartHalf1y.append(rad*np.sin(theta))
    heartHalf2x.append(-rad - rad*np.cos(theta))
    heartHalf2y.append(rad*np.sin(theta))

plt.plot(heartHalf1x,heartHalf1y,color='red', linewidth=4)
plt.plot(heartHalf2x,heartHalf2y,color='red', linewidth=4)
plt.plot([0,heartHalf1x[0]],[ymin,heartHalf1y[0]],color='red', linewidth=4)
plt.plot([0,heartHalf2x[0]],[ymin,heartHalf2y[0]],color='red', linewidth=4)
plt.plot([0,heartHalf2x[0],heartHalf1x[0], 0],[ymin,heartHalf2y[0], heartHalf1y[0], 0],'o',color='black')
plt.grid(True)
plt.gca().set_aspect('equal')
```
